```python
import matplotlib.pyplot as plt
```

```python
def readScores(fname):
    scores = []
    with open(fname) as f:
        lines = f.readlines()
        for line in lines:
            if line == "":
                continue
            scores.append(float(line))
    return scores
```

```python
cartpole1 = readScores('cartpole1.scores')
lunarlander = readScores('lunarlander.scores')
alien = readScores('alien.scores')
asteroid = readScores('asteroid.scores')

print(len(alien), len(asteroid))
```

```python
score_list = [cartpole1, lunarlander, alien, asteroid]
names_list = ['Cart Pole', 'Lunar Lander', 'Alien', 'Asteroid']

for i, item in enumerate(score_list):
    name = names_list[i]
    plt.plot(item)
    plt.title("M-IQN on " + name)
    plt.ylabel('Score')
    plt.xlabel('Episodes')
    plt.savefig(name)
    plt.show()
```
